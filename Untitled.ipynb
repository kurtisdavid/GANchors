```python
def sample_from_big_gan(t_=.75, bs_=4, label_vec_=['husky']):
    model = BigGAN.from_pretrained('biggan-deep-128')
    truncation = t_
    bs = bs_
    class_vector = one_hot_from_names(label_vec_, batch_size=bs)
    noise_vector = truncated_noise_sample(truncation = truncation, batch_size = bs)
    noise_vector = torch.from_numpy(noise_vector)
    class_vector = torch.from_numpy(class_vector)
    
    # GPU
    noise_vec = noise_vector.to('cuda')
    class_vec = class_vector.to('cuda')
    model.to('cuda')
    
    with torch.no_grad():
        output = model(noise_vec, class_vec, truncation)
    output = output.to('cpu')
    
    return {'samples':output,'z_n':noise_vector,'z_c':class_vector}
```
